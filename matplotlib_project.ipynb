```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales_data = pd.read_csv('data.csv')
sales_data.head(5)
```

```python
sales_data.isnull().sum()
```

```python
sales_data["Sales"] = sales_data["Sales"].fillna(sales_data["Sales"].mean())
sales_data["Value"] = sales_data["Value"].fillna(sales_data["Value"].mean())
```

```python
sales_data.isnull().sum()
```

```python
sales_by_region = sales_data.groupby("Region")["Sales"].sum()
sales_by_region
```

```python
sales_by_region.plot(kind="bar",color="pink")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.title("Bar chart for Sales by Region")
plt.show()
```

```python
sales_by_time = sales_data.groupby("Date")["Sales"].sum().reset_index()
sales_by_time
```

```python
plt.figure(figsize=(10,5))
plt.plot(sales_by_time["Date"],sales_by_time["Sales"],color="red")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Bar chart for Sales by Date")
plt.gcf().autofmt_xdate()
plt.show()
```
