```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y= [1,4,6,8,12]
```

```python
# creating a line plot 
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Basic Line Plot")
```

```python
# customize line chart
plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=2.5,markersize=7)
plt.grid(True)
```

```python
x = [1,2,3,4]
y1 = [2,4,6,8]
y2 = [4,6,8,10]

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.plot(x,y1,color="green")
plt.title("Plot 1")

plt.subplot(1,2,2)
plt.plot(y1,y2,color="red")
plt.title("Plot 2")
```

```python
# Bar Plot
categories = ['A','B','C','D','E']
values = [2,5,3,6,1]

plt.bar(categories,values,color="pink")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Graph")
plt.show()
```

```python
# Histogram 
data = [1,2,4,5,5,6,1,4,5,6,3,2,2,2,3,2,2,2,1,1,5,5,5,6,3,2,1]
plt.hist(data,bins=6,color="orange",edgecolor="black")
```

```python
# Scatter Plot
x = [1,3,5,7,9]
y= [2,4,6,8,10]

plt.scatter(x,y,color="red",marker="x")
```

```python
# Pie Chart 
labels = ['A','B','C','D','E']
sizes = [10,20,30,10,30]
colors=['olivedrab', 'rosybrown', 'pink', 'blue','purple']
explode = (0.2,0,0,0,0,)

plt.pie(sizes,explode=explode,labels=labels,colors=colors, autopct='%.0f%%',shadow=True)
plt.show()
```
